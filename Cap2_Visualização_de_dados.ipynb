##Criando gráficos
___

```
import pandas as pd

from matplotlib import pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

```
%matplotlib inline
rcParams['figure.figsize'] = 10, 8
sns.set_style('whitegrid')
```

Criando um gráfico de linha a partir de uma lista

Desenhando um gráfico de linha usando o matplotlib

```
x = range(1, 10)
y = [1, 2, 3, 4, 3, 4, 3, 2, 1]


plt.plot(x, y)
```

Desenhando um gráfico de linha usando um objeto Pandas

```
caminho = '/mtcars.csv'

carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec',
                  'vs', 'am', 'gear', 'carb']
```

```
df = carros[['cyl', 'wt', 'mpg']]
df.plot()
```

##Criando gráficos de barras
___

```
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sns.set_style('whitegrid')
```

Criando gráficos de barras

Desenhando um gráfico de barras a partir de uma lista com matplotlib

```
x = range(1, 10)
y = [1, 2, 3, 4, 3, 4, 3, 2, 1]

plt.bar(x, y)
```

Desenhando um gráfico de barras com um objeto Pandas

```
# caminho = '/mtcars.csv'

carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec',
                  'vs', 'am', 'gear', 'carb']
mpg = carros['mpg']
```

```
mpg.plot(kind='bar')
```
