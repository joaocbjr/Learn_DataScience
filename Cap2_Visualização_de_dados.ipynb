##Gráficos e o Python
___

####Dois métodos para construir gráficos

* **Método funcional**: *com esse método, construímos os gráficos chamando as funções a partir de uma variável ou conjunto de varáveis.*

* **Método orientado ao objeto**: *com esse método, construímos o gráfico primeiro escrevendo uma figura em branco e em seguida populando o gráfico com novos elementos.*

####Gráficos de linha e de barra
**Gráfico de linha**

> Mostra mudanças de valor de um atributoem relaão a uma variável X.

> Pode ser usado para comparar visualmente vários atributos relacionados.

**Gráfico de barra**

> Representa atributos de uma categoria específica usando barras de tamanhos diferentes.

> Representa a contagem de observações dentro de cada categoria.

## Formatando gráficos
___

```
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from pylab import rcParams

import seaborn as sns


%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sns.set_style('whitegrid')
```

Definindo uma cor

```
x = range(1, 10)
y = [1,2,3,4,0.5,4,3,2,1]

plt.bar(x, y)
```

```
cor = ['grey']
plt.bar(x, y, color=cor, align='center')
```

Customizando estilos das linhas

```
x1 = range(0,7)
y1 = [1, 3, 8, 2, 1, 2, 1]

plt.plot(x, y)
plt.plot(x1,y1)
```

```
plt.plot(x, y, ls='steps', lw=5)
plt.plot(x1, y1, ls='--', lw=10)
```


Definindo marcadores em gráficos

```
plt.plot(x, y, marker='1', mew=20)
plt.plot(x1, y1, marker='+', mew=15)
plt.show()
```

##Conhecendo histogramas, box plots e scatter plots
___

**Histograma**
>Mostar a distribuição de uma variável como um conjunto de retângulos adjacentes num gráfico.

>Representa a contagem de dados dentro de um intervalo numérico de valores

**Scatterplot**
>Útil para explorar relações ou dependência entre duas variáveis.

>Ideal para entrar outliers visualmente e também outras tendências.

**Matriz de Scatterpolt**
>Conjunto no formato de matrizes para identificar correlações entre duas ou mais variáveis.

**Boxplot**
>Gráfico para varificar a dispersão dos dados e a presença de outliers





___
##Montando histogramas, box plots e scatter plots
___

```
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
from matplotlib import rcParams
import seaborn as sns

%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sns.set_style('whitegrid')
```

Supondo a distribuição dos dados com histogramas

```
caminho = '/content/mtcars.csv'

carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec',
                  'vs', 'am', 'gear', 'carb']
carros.index = carros.nomes
mpg = carros['mpg']

plt.hist(mpg)
plt.plot()
plt.show()
```

```
sns.distplot(mpg)
```

Vendo os scatterplots em ação

```
carros.plot(kind='scatter', x='hp', y='mpg', c=['darkgray'], s=150)
```

```
sns.regplot(x='hp', y='mpg', data=carros, scatter=True)
```

___
##Gerando uma matriz de scatter plots
___

```
sns.pairplot(carros)
```

```
colunas = ['mpg', 'disp', 'hp', 'wt']
carros_df = pd.DataFrame((carros[colunas].values), columns = colunas)
carros_target = carros['am'].values

carros_df['group'] = pd.Series(carros_target, dtype='category')
```

```
sns.pairplot(carros_df, hue='group', palette='hls')
```

Construindo boxplots

```
carros.boxplot(column='mpg', by='am')
carros.boxplot(column='wt', by='am')
```

```
sns.boxplot(x='am', y='mpg', data=carros, palette='hls')
```
